```python
# 绘制x轴和y轴的刻度
import matplotlib.pyplot as plt
import random
x= range(2 , 26, 2)#最后的2是步长
y= [random.randint (15, 30) for i in x]# x有多少个，y就生成多少个
# 设置图片的大小
plt.figure(figsize= (10, 4), dpi=80)
#设置x轴的刻度标签
x_ticks_label=[ '{}:00' .format(i) for i in x]
plt.xticks(x, x_ticks_label)
#设置y轴的刻度标签
y_ticks_label=[ '{}℃' .format(i) for i in range(min(y), max(y)+1) ]
plt.yticks(range(min(y),max(y)+1), y_ticks_label)
plt.xlabel('time')
plt.ylabel('temperature')
plt.title('diurnal temperature varistion')
plt.plot(x, y)
plt.show()
```

```python
#三月份每一天的最高气温
from matplotlib import pyplot as plt
y = [11, 17, 16, 11, 12, 11, 12, 6, 6, 7, 8, 9, 12, 15, 14, 17, 18, 21, 16, 17, 20, 14, 15, 15, 15, 19, 21, 22, 22, 22, 23]
x = range(1, 32)
#设置图形的大小
plt.figure(figsize=(20,8), dpi= 80)
#使用scatter绘制散点图
plt.scatter(x,y, label='March')
#x轴的刻度
x_ticks_label=['march{}' .format(i) for i in x]
plt.xticks(x[: : 3], x_ticks_label[: : 3])
plt.xlabel('date')
plt.ylabel('temperature')
plt.legend()
plt.show()
```

绘制条形图

```python
import matplotlib.pyplot as plt
a = ['parasite', 'joker', 'avengers:endgame', 'the wandering earth', 'green book', 'spirited away']
b = [38.13, 19.85, 14.89, 11.36, 6.47, 5.93]
#升序的冒泡排序法
def bubble_sort(nums):

    for i in range(len(nums) - 1):  # 这个循环负责设置冒泡排序进行的次数（比如说n个数，则只要进行n-1次冒泡，就可以把这个n个数排序好，对吧）
        for j in range(len(nums) - i - 1):  
            if nums[j] > nums[j + 1]:
                nums[j], nums[j + 1] = nums[j + 1], nums[j]

    return nums

print(bubble_sort((b)))
plt.figure( figsize=(20,8), dpi= 80)
#绘制条形图
rects=plt.bar(range(len(a)), b, width= 0.3, color= 'red')
plt.xticks(range(len(a)), a)
plt.yticks()
#在条形图上加标注（水平居中）
for rect in rects:
    height = rect.get_height()
    plt.text(rect.get_x()+rect.get_width() / 2, height+0.3, str(height), ha="center")
plt.show

```
